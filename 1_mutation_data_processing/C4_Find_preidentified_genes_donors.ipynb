```python
## Find whether patients carry mutations in genes that are known to be regulate mtDNA-CN (results from Gupta et al., 2023)

genelist = {}

glistfile = 'NucGenes_mtCN_Gupta2023.txt'
glfile = open(glistfile,'r')
rdlist = glfile.readlines()
for rd in rdlist:
    rd = rd.strip('\n')
    genelist[rd]=1
    ##print(rd)

glfile.close()

glistfile = None


wrfile = '../results_June2025/MutationGeneList_DonorWise_Mut_mtCNreg_genes.txt'
wrf = open(wrfile,'w') 

resfile = '../results_June2025/MutationGeneList_DonorWise.txt'
rfile = open(resfile,'r')
lines = rfile.readlines()

for line in lines: 
    line = line.strip('\n')
    if line.startswith('Cancer'):
        wrf.write('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format('Cancer','DonorID','NumGenes_mtCNreg','NumGeneStrEff_mtCNreg','NumGeneWeakEff_mtCNreg','NumGeneSilentMut_mtCNreg','NumGeneTranscrEff_mtCNreg','NumPromEff_mtCNreg','NumGenesCdsPromEff_mtCNreg','Genes_mtCNreg','GeneStrEff_mtCNreg','GeneWeakEff_mtCNreg','GeneSilentMut_mtCNreg','GeneTranscrEff_mtCNreg','PromEff_mtCNreg','GenesCdsPromEff_mtCNreg'))
    
    if not line.startswith('Cancer'):
        elems = line.split()
        num = len(elems)
        #print(num)
        
        #genes = elems[num-2].split(',')
        #numgenes = len(genes) - 1
        #proms = elems[num-1].split(',')
        #numproms = len(proms) - 1
        
        genes = elems[2].split(',')
        cnt1 = 0
        tmplist1 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist1 = tmplist1 + gene + ','
                cnt1 += 1
        
        genes = None
        
        genes = elems[3].split(',')
        cnt2 = 0
        tmplist2 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist2 = tmplist2 + gene + ','
                cnt2 += 1
        
        genes = None 
        
        genes = elems[4].split(',')
        cnt3 = 0
        tmplist3 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist3 = tmplist3 + gene + ','
                cnt3 += 1
        
        genes = None 
        
        genes = elems[5].split(',')
        cnt4 = 0
        tmplist4 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist4 = tmplist4 + gene + ','
                cnt4 += 1
        
        genes = None 
        
        genes = elems[6].split(',')
        cnt5 = 0
        tmplist5 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist5 = tmplist5 + gene + ','
                cnt5 += 1
        
        genes = None 
        
        genes = elems[7].split(',')
        cnt6 = 0
        tmplist6 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist6 = tmplist6 + gene + ','
                cnt6 += 1
        
        genes = None 


        genes = elems[8].split(',')
        cnt7 = 0
        tmplist7 = ""
        for gene in genes: 
            if gene in genelist.keys():
                #print(gene)
                tmplist7 = tmplist7 + gene + ','
                cnt7 += 1
        
        genes = None 
        
        if tmplist1 == '':
            tmplist1 = 'X'
        if tmplist2 == '':
            tmplist2 = 'X'
        if tmplist3 == '':
            tmplist3 = 'X'
        if tmplist4 == '':
            tmplist4 = 'X'
        if tmplist5 == '':
            tmplist5 = 'X'
        if tmplist6 == '':
            tmplist6 = 'X'
        if tmplist7 == '':
            tmplist7 = 'X'
        
        
        wrf.write('{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n'.format(elems[0],elems[1],cnt1,cnt2,cnt3,cnt4,cnt5,cnt6,cnt7,tmplist1,tmplist2,tmplist3,tmplist4,tmplist5,tmplist6,tmplist7))
        
        #numgenes = None 
        #numproms = None 
        #break
        
wrf.close()
rfile.close()
resfile = None

print('Done')
```
