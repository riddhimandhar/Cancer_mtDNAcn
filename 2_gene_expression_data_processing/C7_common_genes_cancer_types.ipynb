```python
# Find common genes across cancer types among differentially expressed genes 

canctypelist = ['Biliary-AdenoCA' , 
                 'Bladder-TCC',
    	         'Breast-AdenoCA',
                 'CNS-GBM',
                 'CNS-Oligo',
                 'Cervix-SCC',
                 'ColoRect-AdenoCA',
                 'Head-SCC',
                 'Kidney-ChRCC',
                 'Kidney-RCC',
                 'Liver-HCC',
                 'Lung-AdenoCA',
                 'Lung-SCC',
                 'Lymph-BNHL',
                 'Lymph-CLL',
                 'Ovary-AdenoCA',
                 'Panc-AdenoCA',
                 'Prost-AdenoCA',
                 'SoftTissue-Leiomyo',
                 'SoftTissue-Liposarc',
                 'Stomach-AdenoCA',
                 'Thy-AdenoCA',
                 'Uterus-AdenoCA']


uplist = {}
dnlist = {}
updnlist = {} 

statlist = {}



for canctype in canctypelist:
    print(canctype) 

    arr = [0] * len(canctypelist)
    statlist[canctype] = arr
    
    file1 = '../results_GeneExprAnalysis/DiffExprAnalysis_gene/DESeq_results/' + canctype + '_Upreg_High_vs_Low_mtDNA.txt'

    rdf1 = open(file1) 
    lines = rdf1.readlines() 
    for line in lines:
        if not line.startswith('baseMean'):
            elems = line.split()

            if elems[0] in uplist.keys():
                val = uplist[elems[0]]
                val.append(canctype)
                uplist[elems[0]] = val 
                #print(elems[0],val)
                val = None 
            else:
                uplist[elems[0]] = [canctype]
                
    rdf1.close() 

    file2 = '../results_GeneExprAnalysis/DiffExprAnalysis_gene/DESeq_results/' + canctype + '_Dnreg_High_vs_Low_mtDNA.txt'

    rdf2 = open(file2) 
    lines = rdf2.readlines() 
    for line in lines:
        if not line.startswith('baseMean'):
            elems = line.split()

            if elems[0] in dnlist.keys():
                val = dnlist[elems[0]]
                val.append(canctype)
                dnlist[elems[0]] = val 
                #print(elems[0],val)
                val = None 
            else:
                dnlist[elems[0]] = [canctype]
                
    rdf2.close() 


arr = None 
canctype = None 


wrf1 = open('../results_GeneExprAnalysis/Common_unique_genes/Upreg_common_genes.txt','w')
wrf2 = open('../results_GeneExprAnalysis/Common_unique_genes/Upreg_unique_genes.txt','w')

print("\n\nUp genes")

sg1 = dict(sorted(uplist.items(), key=lambda item: len(item[1]), reverse=True))

uniquelist = {} 

for key in sg1: 
    #print(key,len(uplist[key]),uplist[key])
    if len(uplist[key]) > 1 : 
        wrf1.write('{}\t{}\t{}\n'.format(key,len(uplist[key]),uplist[key]))
    if len(uplist[key]) == 1: 
        wrf2.write('{}\t{}\t{}\n'.format(key,len(uplist[key]),uplist[key]))
        for canctype in uplist[key]:
            #print(canctype)
            if canctype in uniquelist.keys():
                val = uniquelist[canctype]
                val.append(key)
                uniquelist[canctype] = val
                val = None 
            else:
                uniquelist[canctype] = [key]

    ## New part -- cancer specific statistics 
    for canctype in uplist[key]: 
        val = statlist[canctype]
        pos = len(uplist[key])-1
        val[pos] += 1 
        statlist[canctype] = val
        #print(canctype,val)
        pos = None 
        val = None 


wrf1.close()
wrf2.close()

canctype = None 
key = None 
sg1 = None 
uplist = None 
wrf1 = None 
wrf2 = None 
key = None 


## Cancer specific unique gene list 

for key in uniquelist:
    filename = '../results_GeneExprAnalysis/Common_unique_genes/Upreg_unique_lists_canctype/' + key + '_unique_list.txt'
    wrf = open(filename,'w')
    #print(key,uniquelist[key])
    for elem in uniquelist[key]:
        wrf.write('{}\n'.format(elem))
        
    wrf.close()

elem = None 
filename = None 
uniquelist = None 



wrf1 = open('../results_GeneExprAnalysis/Common_unique_genes/Dnreg_common_genes.txt','w')
wrf2 = open('../results_GeneExprAnalysis/Common_unique_genes/Dnreg_unique_genes.txt','w')

print("Dn genes")

sg2 = dict(sorted(dnlist.items(), key=lambda item: len(item[1]), reverse=True))

uniquelist = {} 

for key in sg2: 
    #print(key,len(dnlist[key]),dnlist[key])
    if len(dnlist[key]) > 1 : 
        wrf1.write('{}\t{}\t{}\n'.format(key,len(dnlist[key]),dnlist[key]))
    if len(dnlist[key]) == 1: 
        wrf2.write('{}\t{}\t{}\n'.format(key,len(dnlist[key]),dnlist[key]))
        for canctype in dnlist[key]:
            #print(canctype)
            if canctype in uniquelist.keys():
                val = uniquelist[canctype]
                val.append(key)
                uniquelist[canctype] = val
                val = None 
            else:
                uniquelist[canctype] = [key]
                
    ## New part -- cancer specific statistics 
    for canctype in dnlist[key]: 
        val = statlist[canctype]
        pos = len(dnlist[key])-1
        val[pos] += 1 
        statlist[canctype] = val
        #print(canctype,val)
        pos = None 
        val = None 

wrf1.close()
wrf2.close()

canctype = None 
sg2 = None 
dnlist = None 
key = None 
wrf1 = None 
wrf2 = None 


## Cancer specific unique gene list 

for key in uniquelist:
    filename = '../results_GeneExprAnalysis/Common_unique_genes/Dnreg_unique_lists_canctype/' + key + '_unique_list.txt'
    wrf = open(filename,'w')
    #print(key,uniquelist[key])
    for elem in uniquelist[key]:
        wrf.write('{}\n'.format(elem))
        
    wrf.close()

elem = None 
filename = None 
uniquelist = None 

wrfile = open('../results_GeneExprAnalysis/Common_unique_genes/Statlist_cancerWise_UpDn_shared.txt','w')

for canctype in statlist:
    wrfile.write('{}'.format(canctype))
    for k in range(0,len(statlist[canctype])):
        wrfile.write('\t{}'.format(statlist[canctype][k]))
    wrfile.write('\t{}\n'.format(sum(statlist[canctype])))

wrfile.close()

statlist = None 
```
