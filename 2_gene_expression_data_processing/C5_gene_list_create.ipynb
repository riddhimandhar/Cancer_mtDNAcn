```python
## Create gene lists in the 'genelist' folder 

canctypelist = ['Biliary-AdenoCA',
                'Bladder-TCC',
                'Breast-AdenoCA',
                'CNS-GBM',
                'CNS-Oligo',
                'Cervix-SCC',
                'ColoRect-AdenoCA',
                'Head-SCC',
                'Kidney-ChRCC',
                'Kidney-RCC',
                'Liver-HCC',
                'Lung-AdenoCA',
                'Lung-SCC',
                'Lymph-BNHL',
                'Lymph-CLL',
                'Ovary-AdenoCA',
                'Panc-AdenoCA',
                'Prost-AdenoCA',
                'SoftTissue-Leiomyo',
                'SoftTissue-Liposarc',
                'Stomach-AdenoCA',
                'Thy-AdenoCA',
                'Uterus-AdenoCA']


path = '../results_GeneExprAnalysis/DiffExprAnalysis_gene/DESeq_results/'

for canc in canctypelist: 
    print(canc)
    file1 = path + canc + '_Dnreg_High_vs_Low_mtDNA.txt'
    file2 = path + canc + '_Upreg_High_vs_Low_mtDNA.txt'
    wrfile1 = path + 'genelist/' + canc + '_Dnreg.txt'
    wrfile2 = path + 'genelist/' + canc + '_Upreg.txt'
    
    #print(file1)
    wrf1 = open(wrfile1,'w')
    rdfile = open(file1)
    lines = rdfile.readlines() 
    
    for line in lines: 
        line = line.rstrip('\n')
        if not line.startswith('baseMean'):
            elems = line.split()
            wrf1.write('{}\n'.format(elems[0]))
            elems = None 
     
    rdfile.close()
    wrf1.close()
        
    line = None 
    lines = None 
    wrf1 = None 
    rdfile = None 
    
    
    wrf2 = open(wrfile2,'w')
    rdfile = open(file2)
    lines = rdfile.readlines() 
    
    for line in lines: 
        line = line.rstrip('\n')
        if not line.startswith('baseMean'):
            elems = line.split()
            wrf2.write('{}\n'.format(elems[0]))
            elems = None 
     
    rdfile.close()
    wrf2.close()
        
    line = None 
    lines = None 
    wrf2 = None 
    rdfile = None 
    

    file1 = None 
    file2 = None 
    
canctypelist = None 
```
