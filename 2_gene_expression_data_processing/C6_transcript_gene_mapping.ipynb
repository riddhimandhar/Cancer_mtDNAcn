```python
## Map transcripts to Gene
## Gather statistics 

maplist = {}

mapfile = open('../../../datasets/gene_protein_name_mapping/Homo_sapiens_mapping.txt')

lines = mapfile.readlines()

for line in lines:
    elems = line.split()
    maplist[elems[1]] = elems[3]

mapfile.close()

line = None 
lines = None 


canctypelist = ['Biliary-AdenoCA',
                'Bladder-TCC',
                'Breast-AdenoCA',
                'CNS-GBM',
                'CNS-Oligo',
                'Cervix-SCC',
                'ColoRect-AdenoCA',
                'Head-SCC',
                'Kidney-ChRCC',
                'Kidney-RCC',
                'Liver-HCC',
                'Lung-AdenoCA',
                'Lung-SCC',
                'Lymph-BNHL',
                'Lymph-CLL',
                'Ovary-AdenoCA',
                'Panc-AdenoCA',
                'Prost-AdenoCA',
                'SoftTissue-Leiomyo',
                'SoftTissue-Liposarc',
                'Stomach-AdenoCA',
                'Thy-AdenoCA',
                'Uterus-AdenoCA']


wrf = open('../results_GeneExprAnalysis/DiffExprAnalysis_transcript/Stats_Genelevel_Up_Down_reg.txt','w')

path = '../results_GeneExprAnalysis/DiffExprAnalysis_transcript/DESeq_results/'
path2 = '../results_GeneExprAnalysis/DiffExprAnalysis_transcript/DESeq_results_genes/'

for canc in canctypelist: 
    print(canc)
    wrf.write('{} '.format(canc))
    file1 = path + canc + '_Dnreg_High_vs_Low_mtDNA.txt'
    file2 = path + canc + '_Upreg_High_vs_Low_mtDNA.txt'
    file3 = path + canc + '_High_vs_Low_mtDNA.txt'
    wrfile1 = path2 + canc + '_Dnreg_genes_High_vs_Low_mtDNA.txt'
    wrfile2 = path2 + canc + '_Upreg_genes_High_vs_Low_mtDNA.txt'  
    wrall = path2 + canc + '_genes_High_vs_Low_mtDNA.txt'  
    
    #print(file1)
    wrf1 = open(wrfile1,'w')
    rdfile = open(file1)
    lines = rdfile.readlines() 
    
    dnlist = {}
    
    for line in lines: 
        line = line.rstrip('\n')
        #print(line,'Gene')
        if line.startswith('baseMean'):
            wrf1.write('{} '.format(line))
            wrf1.write('Gene\n')
        if not line.startswith('baseMean'):
            elems = line.split()
            if elems[0] in maplist.keys():
                wrf1.write('{} '.format(line))
                wrf1.write('{}\n'.format(maplist[elems[0]]))
                dnlist[maplist[elems[0]]] = 1
            #if not elems[0] in maplist.keys():
            #    wrf1.write('{}\n'.format(elems[0]))
            #    dnlist[elems[0]] = 1
            elems = None 
     
    rdfile.close()
    wrf1.close()
        
    line = None 
    lines = None 
    wrf1 = None 
    rdfile = None 
    
    
    uplist = {}
    
    wrf2 = open(wrfile2,'w')
    rdfile = open(file2)
    lines = rdfile.readlines() 
    
    for line in lines: 
        line = line.rstrip('\n')
       
        if line.startswith('baseMean'):
            wrf2.write('{} '.format(line))
            wrf2.write('Gene\n')
        if not line.startswith('baseMean'):
            elems = line.split()
            if elems[0] in maplist.keys():
                wrf2.write('{} '.format(line))
                wrf2.write('{}\n'.format(maplist[elems[0]]))
                uplist[maplist[elems[0]]] = 1
            #if not elems[0] in maplist.keys():
            #    wrf2.write('{}\n'.format(elems[0]))
            #    uplist[elems[0]] = 1
            elems = None 
     
    rdfile.close()
    wrf2.close()
        
    line = None 
    lines = None 
    wrf2 = None 
    rdfile = None 
    
    wrf.write('{}:{}\t{}:{}\n'.format('Upreg_genes',len(uplist.keys()),'Dnreg_genes',len(dnlist.keys())))
    
    wrfile3 = path2 + 'genelist/' + canc + '_Dnreg.txt'
    wrfile4 = path2 + 'genelist/' + canc + '_Upreg.txt'
    
    wrf3 = open(wrfile3,'w')
    for key in dnlist.keys():
        wrf3.write('{}\n'.format(key))
    wrf3.close() 
    
    wrf4 = open(wrfile4,'w')
    for key in uplist.keys():
        wrf4.write('{}\n'.format(key))
    wrf4.close() 
        
    uplist = None 
    dnlist = None 


    # All transcripts
    wrf_full = open(wrall,'w')
    rdfile = open(file3)
    lines = rdfile.readlines() 
      
    for line in lines: 
        line = line.rstrip('\n')
        if line.startswith('baseMean'):
            wrf_full.write('{} '.format(line))
            wrf_full.write('Gene\n')
        if not line.startswith('baseMean'):
            elems = line.split()
            if elems[0] in maplist.keys():
                wrf_full.write('{} '.format(line))
                wrf_full.write('{}\n'.format(maplist[elems[0]]))

            elems = None 
     
    rdfile.close()
    wrf_full.close()
        
    line = None 
    lines = None 
    wrf_full = None 
    rdfile = None 
    

wrf.close()

file1 = None 
file2 = None 
canctypelist = None 
maplist = None 
```
