```python
## Alternative splicing 

import pandas as pd 

trpath = '../results_GeneExprAnalysis/DiffExprAnalysis_transcript/DESeq_results_genes/'

wrpath = '../results_GeneExprAnalysis/Comparison_diffExprAnalysis_gene_transcript/alternative_splicing/'

canctypelist = ['Biliary-AdenoCA', 
                 'Bladder-TCC',
    	         'Breast-AdenoCA',
                 'CNS-GBM',
                 'CNS-Oligo',
                 'Cervix-SCC',
                 'ColoRect-AdenoCA',
                 'Head-SCC',
                 'Kidney-ChRCC',
                 'Kidney-RCC',
                 'Liver-HCC',
                 'Lung-AdenoCA',
                 'Lung-SCC',
                 'Lymph-BNHL',
                 'Lymph-CLL',
                 'Ovary-AdenoCA',
                 'Panc-AdenoCA',
                 'Prost-AdenoCA',
                 'SoftTissue-Leiomyo',
                 'SoftTissue-Liposarc',
                 'Stomach-AdenoCA',
                 'Thy-AdenoCA',
                 'Uterus-AdenoCA']

wrf = open('../results_GeneExprAnalysis/Comparison_diffExprAnalysis_gene_transcript/alternative_splicing/Stats_canctype.txt','w')
wrf.write('Canctype\tUptr_genes\tDntr_genes\tNum_both_UpDntr\n') 

for canc in canctypelist:
    print(canc)
    uptr = trpath + canc + '_Upreg_genes_High_vs_Low_mtDNA.txt'
    dntr = trpath + canc + '_Dnreg_genes_High_vs_Low_mtDNA.txt'

    uptr_data = pd.read_csv(uptr,sep=' ')
    uptr_genes = list(set(list(uptr_data.loc[:,'Gene'])))

    dntr_data = pd.read_csv(dntr,sep=' ')
    dntr_genes = list(set(list(dntr_data.loc[:,'Gene'])))

    common_uptr_dntr = list(set(uptr_genes) & set(dntr_genes))
    #print(len(uptr_genes),len(dntr_genes),len(common_uptr_dntr))
    wrf.write('{}\t{}\t{}\t{}\n'.format(canc,len(uptr_genes),len(dntr_genes),len(common_uptr_dntr)))

    df_common = pd.DataFrame(common_uptr_dntr, columns = ['Gene'])
    gnfile = '../results_GeneExprAnalysis/Comparison_diffExprAnalysis_gene_transcript/alternative_splicing/genelist/' + canc + '.txt'
    df_common.to_csv(gnfile, header=False, index=False)
    gnfile = None 
    
    if len(common_uptr_dntr) != 0: 
        ## print transcript expressions 
        wrfile = wrpath + 'cancerWise/' + canc + '_alt_spliced_transcripts.txt' 
        cwr = open(wrfile,'w')

        #print(uptr_data)
        filt_up_rows = uptr_data.loc[uptr_data['Gene'].isin(common_uptr_dntr)] 
        #print(filt_up_rows)

        filt_dn_rows = dntr_data.loc[dntr_data['Gene'].isin(common_uptr_dntr)] 
        #print(filt_dn_rows)

        val = pd.concat([filt_up_rows,filt_dn_rows],axis=0)
        val_sorted = val.sort_values(by='Gene')
        #print(val_sorted)
        val_sorted.to_csv(wrfile,sep='\t')
        cwr.close()

        wrfile = None 
        filt_up_rows = None 
        filt_dn_rows = None 
        val = None 
        val_sorted = None 
    
    uptr_data = None 
    uptr_genes = None 
    dntr_data = None 
    dntr_genes = None 
    
wrf.close() 

canctypelist = None 
```
